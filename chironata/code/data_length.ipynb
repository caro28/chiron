```python
jowett_tokenized = []
for sent in jowett_lst:
    jowett_tokenized.append(sent.split())

lengths_lst = []
for token_sent in jowett_tokenized:
    lengths_lst.append(len(token_sent))

all_lengths = 0
for num in lengths_lst:
    all_lengths += num

avg = all_lengths/len(jowett_lst)
avg
```
