```python
# duplicates: deleted values for "WORK IDs" to avoid double-counting later on
path = "/Users/carolinecraig/Documents/Northeastern/translation_project/summer2023/translations_repositories.csv"
translations_repos = pd.read_csv(path, dtype=str)
```

```python
translations_repos.rename(columns={'Author Name':'Author_Name', 'WORK IDs':'WORK_IDs', 
                                   '# of translations':'#_of_translations', 'Edition Title':'Edition_Title'},
                          inplace=True)
```

```python
transl_repos_bywork = (translations_repos.assign(CTS_URNs_exploded=translations_repos["CTS-URN/s"].str.split()).
        explode("CTS_URNs_exploded").
        reset_index(drop=True))
```

```python
transl_repos_bywork.head()
```

```python
transl_repos_bywork.shape
```

```python
# split CTS_URNs to get unique code per work (i.e. remove ending translation language code)
# keep "." before "ogl" and "opp" so that unique codes do not end with digit (for checking)
transl_repos_bywork["CTS_URN_work"] = transl_repos_bywork["CTS_URNs_exploded"].str.split("ogl").str[0]
transl_repos_bywork["CTS_URN_work"] = transl_repos_bywork["CTS_URN_work"].str.split("opp").str[0]
```

```python
transl_repos_bywork["Translation Language"].unique()
```

```python
# check that all CTS_URN_work are unique (i.e. not missing any other conventions - ogl, opp - for splitting)
count = 0
test = transl_repos_bywork["CTS_URN_work"].fillna("None")
for idx, item in enumerate(test):
    if item[-1].isdigit():
        count += 1
        print(idx)
count
```

```python
translations_sorted_by_count = transl_repos_bywork["CTS_URN_work"].value_counts().to_frame().reset_index()
```

```python
translations_sorted_by_count.rename(columns={"count":"translations_count"}, inplace=True)
```

```python
# add work title and author name
work_titles = []
author_names = []
for item in translations_sorted_by_count["CTS_URN_work"]:
    try:
        work_titles.append(transl_repos_bywork.loc[(transl_repos_bywork["CTS_URN_work"]==item) & 
                                (transl_repos_bywork["#_of_translations"]=="1")
                               ]["Edition_Title"].reset_index().iloc[0]["Edition_Title"])
        
        ######## run code block below if want titles in English. Con: will require more manual checking #######
#         work_titles.append(transl_repos_bywork.loc[(transl_repos_bywork["CTS_URNs_unique"]==item) & 
#                                 (transl_repos_bywork["#_of_translations"]=="1") &
#                                                    (transl_repos_bywork["Translation Language"]=="English")
#                                ]["Edition_Title"].reset_index().iloc[0]["Edition_Title"])

    except IndexError:
        work_titles.append("Check")
    
    author_names.append(transl_repos_bywork.loc[transl_repos_bywork["CTS_URN_work"]==item][
            "Author_Name"].iloc[0])
```

```python
# add edition title (book title)
edition_titles = []
for item in translations_sorted_by_count["CTS_URN_work"]:
    edition_titles.append(transl_repos_bywork.loc[transl_repos_bywork[
        "CTS_URN_work"]==item]["Edition_Title"].iloc[0])
```

```python
# add languages
languages = []
translation_languages = []
for item in translations_sorted_by_count["CTS_URN_work"]:
    translation_languages.append(transl_repos_bywork.loc[
        transl_repos_bywork["CTS_URN_work"]==item]["Translation Language"].unique())
```

```python
translations_sorted_by_count["work_title"] = work_titles
translations_sorted_by_count["author_name"] = author_names
translations_sorted_by_count["edition_title"] = edition_titles
translations_sorted_by_count["translation_language"] = translation_languages
```
